# A dataset on plant locations

This dataset contains location information for the plant *Begonia conchifolia*. 

```python
import pandas as pd
begonia=pd.read_csv('../Datasets/B_conch_loc.csv')
begonia
```

```python
begonia=pd.read_csv('../Datasets/B_conch_loc.csv', header=None, names=['a','b','c','d','e','f','g','h','i'])
begonia.head()
```

Check the data types of each variable.

```python
begonia.dtypes
```

```python
# the headers were renamed since there are unknown white spaces between the original numeric column headers
print(begonia.columns.tolist())
```

A new column called 'Reference' which concatenates all the information from columns 4 - 8 is made. Use the code `.astype(str)` if the variables are not coded as strings.

Now drop columns 4 - 8 as we don't need them any more.

```python
begonia['Reference']=begonia['e'].astype(str)+','+begonia['f'].astype(str)+','+begonia['g'].astype(str)+','+begonia['h'].astype(str)+','+begonia['i'].astype(str)
begonia
```

```python
begonia.drop(columns=['e', 'f', 'g', 'h', 'i'], inplace=True)
begonia

#Now let's split column 3 into longitude and latitude. Use `str.split` to split this column and form two new columns called Longitude and Latitude. Note that the separator is ', ' this time.
```

```python
begonia[['Longitude','Latitude']]=begonia['d'].str.split(',',expand=True)
begonia
```

```python
begonia.drop(columns=['d'], inplace=True)
begonia
```

Now drop column 3.

And finally, rename the columns using `['Species_name', 'Country', 'Location', 'Reference', 'Longitude', 'Latitude']`.

```python
begonia.columns=['Species_name','Country','Location','Reference','Longitude','Latitude']
begonia
```
