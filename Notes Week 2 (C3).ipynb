# **Tidy Data**

#### Every value belongs to a variable and an observation.
#### a) A value is a single piece of data
###### Each value must have its own cell.
#### b) A variable contains all values that measure the same underlying attribute
###### Each variable must have its own column.
#### c) An observation contains all values measured on the same unit
###### Each observation must have its own row.

### 1) Adding and Removing Data
##### a) Add column using df['COLUMN NAME']=[LIST] or df['NEW COLUMN NAME']=df['COLUMN NAME']/10 or any mathematical operation like in Excel.
##### b) Remove column using df.drop(columns=' COLUMN NAME'), if more than one use df.drop(columns = ['COLUMN_NAME1', 'COLUMN_NAME2']) or df.drop ('COLUMN_NAME1' and 'COLUMN_NAME2')

##### remove row using df.dropdf.drop(ROW_NUMBER). use df.drop(5) to drop 5th row

```python
import pandas as pd
df = pd.read_csv('../Datasets/count.csv')
df['Soil'] = ['Sand', 'Loam', 'Loam', 'Clay', 'Clay', 'Loam', 'Sand', 'Sand', 'Clay', 'Clay']
df['Drainage'] = ['Good', 'OK', 'Poor', 'Poor', 'Poor', 'Good', 'Good', 'OK', 'OK', 'Poor']
df
```

```python
df['Cereals']=df['Barley']+df['Oats']
df
```

```python
df= df.drop(columns= 'Cereals')
df
```

### 2) Rearranging Data (always remember that column starts with [0])
##### a) Listing columns, use var= df.columns.tolist()
##### b) rearranging columns, var_new= [var[1], var[5], var[3]....]
##### c) columns in (alphabetical/numerical) ascending order, use var_sorted= sorted(var)
#####    row ascending, use df=df.sort_values(by='COLUMN NAME', ascending= True)
##### d) columns in reverse order, use var_reversed=sorted(var, reverse=True)
#####    row descending, code ascending=False

```python
cols= df.columns.tolist()
print(cols)
cols_new= [cols[0],cols[6],cols[1],cols[2],cols[3],cols[5],cols[4]]
print(cols_new)
cols_sorted= sorted(cols)
print (cols_sorted)
cols_reversed=sorted(cols, reverse=True)
print(cols_reversed)
print()
df=df.sort_values(by='Oats', ascending=True)
print(df)
print()
df=df.sort_values(by='Barley', ascending=False)
print (df)
```

### 3) Concatenation & Split
##### a) concatenate string columns, df[NEW_COLUMN_NAME]= df['COLUMN_NAME']+'_'+df[COLUMN_NAME]......
##### b) concatenate string and non string columns, df['NEW_COLUMN_NAME']=df['STRING COLUMN NAME']+'_'+ df['INTEGER COLUMN NAME'].astype(str)
##### c) concatenate non string columns, df['NEW_COLUMN_NAME']=df['INTEGER COLUMN NAME'].astype(str)+'_'+ df['INTEGER COLUMN NAME'].astype(str)
##### Notes: in concatenation, '_'is optional for spacing 
##### d) split columns, df[['Soil_new', 'Drainage_new']] = df['CONCATENATED1_CONCATENATED2'].str.split('_', expand = True)

```python
df['Field_Drainage']=df['Field']+'_'+df['Drainage']
df['Barley_Soil_Pair']=df['Barley'].astype(str)+'_'+df['Soil']
df['Goats_Barley_Oats']=df['Goats'].astype(str)+'/'+df['Barley'].astype(str)+'/'+df['Oats'].astype(str)
df
```

```python
df[['Barley_New','Soil_New']]=df['Barley_Soil_Pair'].str.split('_',expand=True)
df
```

##### b) reset index after removal, add .reset_index(drop=True)

```python
df_other = pd.DataFrame([[1, 6.5, 3], [1, 6.5, 3], [2, 4, 7.2], [0.9, 6.5, 3]])
df_other
```
