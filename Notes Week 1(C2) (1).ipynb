##### **Example 1**

```python
# We now will see the uncleaned data frame and troubleshoot it

df_duff1= pd.read_table('../Datasets/count_duff1.txt')
df_duff1.head()

# table below represents missing datas as NaN and *, which will cause problems during calculations
```

```python
df_duff1.info()
df_duff1.dtypes
```

```python
# notice that the data frame above has float64 and object, this shows inconsistent/uncleaned data
# Now we will clean the data thru (1)standardising followed by (2)replacing
# (1) na_values = "*" tells Python to also interpret * as a missing value, so that our * has been turned to NaN (which is the standard way to interpret missing values). 
# (2) dtype specifies data type, pd.Int64Dtype() allows integer columns to hold missing values without converting the whole column to float. 
# result: Missing values are represented as <NA> instead of NaN or *.

df_duff1_take3 = pd.read_table("../Datasets/count_duff1.txt", na_values = "*", dtype = pd.Int64Dtype())
df_duff1_take3.head() 
```

```python
# all dtype has now become int64
df_duff1_take3.info()
```

### **TIDYING DATA**

```python
df_AUK=pd.read_excel('../Datasets/AUK-Chapter5-13jul23.xlsx')
df_AUK
```

##### 1) untidy data with rows with no values at all at its head and tail. remove these rows with skiprows=... and skipfooter=...

```python
# in this example, first 8 rows and last 7 rows have no value

df_AUK= pd.read_excel ('../Datasets/AUK-Chapter5-13jul23.xlsx', skiprows= 8, skipfooter =7)
df_AUK
```

##### 2) untidy data with few rows have missing values. remove these rows using df.dropna() or df.dropna(inplace=True)

```python
- df.dropna() removes any row that contains a missing value (NaN or <NA>).
By default, it does not change the original DataFrame, but instead returns a new DataFrame without the missing rows
- df.dropna.(inplace=True) modifies the df directly instead of keeping the original a copy
```

```python
df_AUK.dropna(inplace=True)
df_AUK
```

```python
There are lots of ways you can adapt this method, for example...

(1) df.dropna(axis = 'columns', inplace = True) drops the columns containing missing values
(2) df.dropna(thresh = 2, inplace = True) drops the rows with more than two missing values
(3) df.dropna(subset = [1983, 1997], inplace = True) drops rows with missing values in the '1983' or '1997' columns
```

```python
Explanations:
axis=0 or by default, the command df.dropna() acts on rows. thats why number (2) doesnt have axis='rows'
axis=1 or axis='columns' to act on column. but if want to remove 2 columns, put df.dropna(axis='columns', thresh=2)
```
